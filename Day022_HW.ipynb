```python
#查詢台北市士林區 2018/01 – 2018/08 的 SO2 資料
browser = webdriver.Chrome(executable_path)
browser.get("http://taqm.epa.gov.tw/taqm/tw/MonthlyAverage.aspx")
selectYear = Select(browser.find_element_by_id("ctl05_ddlYear"))
selectYear.select_by_value('2018')
browser.find_element_by_id('ctl05_btnQuery').click()
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
t = {}
for tr in table.find_all('tr')[1:]:

    for i, td in enumerate(tr.find_all('td')):
        
        if len(tr.find_all('td')) == 5:
            if i == 0:
                obs = td.text
                t.setdefault(obs, {})
            if i == 2:
                date = td.text
            if i == 3:
                value = td.text
                t[obs][date] = value
        
        if len(tr.find_all('td')) == 3:
            if i == 0:
                date = td.text
            if i == 1:
                value = td.text
                t[obs][date] = value
print(t)
```

```python
print("台北市士林區 2018 的 SO2 資料")
print(t['SO2'])
print("台北市士林區 2018/01 – 2018/08 的 SO2 資料")
#參考查詢方式
print(dict([(key, value) for key, value in t['SO2'].items() if key < '2018/09' ]))
```

```python
#參考寫法
import pandas as pd
F=[{'時間': key, '數值': value} for key, value in t['SO2'].items() if key < '2018/09' ]
pd.DataFrame(F)
```

```python
#台北市士林區 2018/01 – 2018/08 的 SO2、CO 資料
d_SO2 = [{'時間': key, 'SO2': value} for key, value in t['SO2'].items() if key < '2018/09' ]
d_CO = [{'時間': key, 'CO': value} for key, value in t['CO'].items() if key < '2018/09' ]
pd.merge(pd.DataFrame(d_SO2), pd.DataFrame(d_CO))
```
